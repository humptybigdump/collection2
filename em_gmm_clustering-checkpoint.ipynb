# EM Example: Gaussian Mixture Fitting

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
```

## Data generation and Model Initialization

```python
# Ground truth data distribution for generation
weights_gth = [0.05, 0.15, 0.3, 0.5]  # Skewed mixture weights
means_gth = [(-5, -5), (0, 0), (5, 5), (10, -5)]  # Means distributed irregularly
scale_factor = 3
covariances_gth = [
    scale_factor * np.array([[2, 0.3], [0.3, 0.5]]),  # Anisotropic covariance
    scale_factor * np.array([[0.5, 0.1], [0.1, 0.7]]),  # Slightly elliptical
    scale_factor * np.eye(2) * 0.8,  # Isotropic covariance
    scale_factor * np.array([[1.5, -0.4], [-0.4, 0.9]])  # Skewed
]
```

```python
def generate_data(n_samples=100):
    n_components = len(weights_gth)
    assert len(means_gth) == n_components and len(covariances_gth) == n_components, \
        "Number of weights, means, and covariances must match the number of components."
    
    # Generate component assignments for each sample
    component_assignments = np.random.choice(
        range(n_components), size=n_samples, p=weights_gth
    )
    
    # Generate samples from the assigned Gaussian components
    X = []
    for i in range(n_components):
        n_samples_component = np.sum(component_assignments == i)
        if n_samples_component > 0:
            samples = np.random.multivariate_normal(
                mean=means_gth[i], cov=covariances_gth[i], size=n_samples_component
            )
            X.append(samples)
    
    X = np.vstack(X)
    return X
```

```python
# fix seed for deterministic action
np.random.seed(42)
X = generate_data(1000)
```

```python
def plot(X, weights, means, covariances):
    """
    Plot a 2D dataset and a GMM defined by weights, means, and covariances.

    Parameters:
    - X (ndarray): The dataset of shape (n_samples, 2).
    - weights (list of float): Mixture weights for each Gaussian component.
    - means (list of tuple): Means of the Gaussian components.
    - covariances (list of ndarray): Covariance matrices of the Gaussian components.
    - labels (ndarray, optional): True labels for each sample (used for coloring).
    """
    plt.figure(figsize=(5, 5))
    
    # Plot the data points
    plt.scatter(X[:, 0], X[:, 1], s=10, color='blue', alpha=1.0)
    
    # Plot the Gaussian components
    for i, (weight, mean, cov) in enumerate(zip(weights, means, covariances)):
        # Plot the mean as an 'x'
        plt.scatter(mean[0], mean[1], color='red', marker='x', s=100, alpha=weight, label=f'Component {i+1}')
        
        # Calculate the ellipse for the covariance
        eigenvalues, eigenvectors = np.linalg.eigh(cov)
        angle = np.arctan2(eigenvectors[1, 0], eigenvectors[0, 0]) * (180 / np.pi)
        width, height = 4 * np.sqrt(eigenvalues)  # Scaled to represent 2 std dev
        
        ellipsis = Ellipse(
            xy=mean, width=width, height=height, angle=angle,
            facecolor='red', alpha=weight *0.8, edgecolor='black'
        )
        plt.gca().add_patch(ellipsis)
    
    plt.title("GMM Visualization")
    plt.xlabel("X1")
    plt.ylabel("X2")
    plt.axis("equal")
    plt.legend()
    plt.show()
```

### Ground Truth data distribution

```python
plot(X, weights_gth, means_gth, covariances_gth)
```

```python
def initialize_gmm_model(n_components, prior_mean_range, prior_cov_scale, n_features=2):
    """
    Initialize a GMM model with random means and large covariance values.

    Parameters:
    - n_components (int): Number of Gaussian components.
    - prior_mean_range (tuple): Range (min, max) for sampling means from a uniform prior distribution.
    - prior_cov_scale (float): Scale factor for generating large covariance values.
    - n_features (int): Number of features (default is 2 for 2D).

    Returns:
    - weights (ndarray): Initial weights for the GMM components, uniformly distributed.
    - means (ndarray): Initial means sampled from the prior distribution.
    - covariances (list of ndarray): Initial large covariance matrices for each component.
    """
    # Initialize weights uniformly
    weights = np.full(n_components, 1 / n_components)
    
    # Sample means from a uniform distribution within the prior range
    means = np.random.uniform(
        low=prior_mean_range[0], high=prior_mean_range[1], size=(n_components, n_features)
    )
    
    # Initialize covariances as scaled identity matrices
    covariances = [np.eye(n_features) * prior_cov_scale for _ in range(n_components)]
    
    return weights, means, covariances
```

```python
n_components = 4
prior_mean_range = (-10, 20)  # Means sampled uniformly in this range
prior_cov_scale = 20  # Large covariance scale

weights, means, covariances = initialize_gmm_model(n_components, prior_mean_range, prior_cov_scale)
```

### Initial Model Distribution

```python
plot(X, weights, means, covariances)
```

## EM Algorithm

## Likelihood Computation

![Image Description](log_likelihood.png)

```python
def multivariate_gaussian_pdf(x, mean, cov):
    """
    Compute the multivariate Gaussian probability density function (PDF).
    
    Parameters:
    - x (ndarray): The data point (1D array).
    - mean (ndarray): The mean of the Gaussian (1D array).
    - cov (ndarray): The covariance matrix (2D array).
    
    Returns:
    - pdf (float): The computed PDF value.
    """
    d = len(mean)  # Dimensionality
    cov_inv = np.linalg.inv(cov)
    cov_det = np.linalg.det(cov)
    
    norm_factor = 1 / np.sqrt((2 * np.pi) ** d * cov_det)
    diff = x - mean
    exponent = -0.5 * diff.T @ cov_inv @ diff
    
    return norm_factor * np.exp(exponent)

def compute_log_likelihood(X, weights, means, covariances):
    """
    Compute the log-likelihood of samples under a Gaussian Mixture Model (GMM).

    Parameters:
    - X (ndarray): Dataset of shape (n_samples, n_features).
    - weights (ndarray): Mixing weights of the GMM components.
    - means (ndarray): Means of the GMM components.
    - covariances (list of ndarray): Covariance matrices of the GMM components.

    Returns:
    - log_likelihood (float): Total log-likelihood of the dataset.
    """
    n_samples, n_features = X.shape
    n_components = len(weights)
    
    # Initialize log-likelihood
    log_likelihood = 0.0
    
    # Loop over all samples
    for x in X:
        # Compute the weighted sum of probabilities for this sample
        p_x = 0.0
        for k in range(n_components):
            p_x += weights[k] * multivariate_gaussian_pdf(x, mean=means[k], cov=covariances[k])
        
        # Add the log of the probability to the log-likelihood
        log_likelihood += np.log(p_x)
    
    return log_likelihood
```

## E-Step

![Image Description](e_step.png)

```python
def compute_responsibilities(X, weights, means, covariances):
    """
    Compute the responsibilities (q_ik) for each data point and each component in the GMM.
    
    Parameters:
    - X (ndarray): Dataset of shape (n_samples, n_features).
    - weights (ndarray): Mixing weights of the GMM components.
    - means (ndarray): Means of the GMM components.
    - covariances (list of ndarray): Covariance matrices of the GMM components.
    
    Returns:
    - responsibilities (ndarray): Matrix of responsibilities with shape (n_samples, n_components).
    """
    n_samples, n_features = X.shape
    n_components = len(weights)
    
    responsibilities = np.zeros((n_samples, n_components))  # Initialize responsibility matrix
    
    for i, x in enumerate(X):
        # Compute the numerator: π_k * N(x_i | μ_k, Σ_k)
        numerators = np.array([
            weights[k] * multivariate_gaussian_pdf(x, means[k], covariances[k])
            for k in range(n_components)
        ])
        
        # Compute the denominator: ∑_j π_j * N(x_i | μ_j, Σ_j)
        denominator = np.sum(numerators)
        
        # Compute responsibilities for each component
        responsibilities[i, :] = numerators / denominator
    
    return responsibilities
```

## Lower Bound Computation

![Image Description](lower_bound.png)

```python
def compute_lower_bound(X, responsibilities, weights, means, covariances):
    """
    Compute the lower bound (ELBO) for the GMM with the given responsibilities.

    Parameters:
    - X (ndarray): Dataset of shape (n_samples, n_features).
    - responsibilities (ndarray): Responsibilities (q_ik) from the E-step.
    - weights (ndarray): Mixing weights of the GMM components.
    - means (ndarray): Means of the GMM components.
    - covariances (list of ndarray): Covariance matrices of the GMM components.

    Returns:
    - lower_bound (float): The computed lower bound (ELBO).
    """
    n_samples, n_features = X.shape
    n_components = len(weights)
    
    # First term: Expected log likelihood (E_q[log p(x_i, z_i)])
    log_likelihood_term = 0.0
    for i, x in enumerate(X):
        # Compute the weighted sum of the log-likelihoods for this sample
        weighted_log_likelihood = 0.0
        for k in range(n_components):
            # Compute the log of the Gaussian PDF (without the normalization constant)
            gaussian_log_pdf = np.log(weights[k]) + np.log(multivariate_gaussian_pdf(x, means[k], covariances[k]))
            # expectation over q_ik: weighting with probability q_ik
            weighted_log_likelihood += responsibilities[i, k] * gaussian_log_pdf
        log_likelihood_term += weighted_log_likelihood
    
    # Second term: Expected log responsibility (E_q[log q_i(z_i)])
    log_responsibility_term = 0.0
    for i in range(n_samples):
        for k in range(n_components):
            if responsibilities[i, k] > 0:
                log_responsibility_term += responsibilities[i, k] * np.log(responsibilities[i, k])
    
    # The final lower bound is the difference between the two terms
    lower_bound = log_likelihood_term - log_responsibility_term
    return lower_bound


```

## M-Step

![Image Description](m_step.png)

```python
def m_step(X, responsibilities):
    """
    Perform the M-step of the EM algorithm for a GMM. This step updates the model parameters
    (weights, means, covariances) based on the computed responsibilities from the E-step.

    Parameters:
    - X (ndarray): Dataset of shape (n_samples, n_features).
    - responsibilities (ndarray): Responsibilities (q_ik) from the E-step.

    Returns:
    - updated_weights (ndarray): Updated mixing weights.
    - updated_means (ndarray): Updated component means.
    - updated_covariances (list of ndarray): Updated covariance matrices for each component.
    """
    n_samples, n_features = X.shape
    n_components = responsibilities.shape[1]
    
    updated_weights = np.zeros(n_components)
    updated_means = np.zeros((n_components, n_features))
    updated_covariances = []

    for k in range(n_components):
        # Compute the total responsibility for component k
        responsibility_sum = np.sum(responsibilities[:, k])

        # Update the mixing weight for component k
        updated_weights[k] = responsibility_sum / n_samples

        # Update the mean for component k
        weighted_sum = np.sum(responsibilities[:, k][:, np.newaxis] * X, axis=0)
        updated_means[k] = weighted_sum / responsibility_sum

        # Update the covariance for component k
        diff = X - updated_means[k]
        weighted_covariance = np.dot((responsibilities[:, k][:, np.newaxis] * diff).T, diff)
        updated_covariances.append(weighted_covariance / responsibility_sum)

    return updated_weights, updated_means, updated_covariances
```

# Execution of the Algorithm

```python
log_likelihoods = []
lower_bounds = []
# Example usage
num_em_iterations = 16
for e in range(num_em_iterations):
    log_likelihood = compute_log_likelihood(X, weights, means, covariances)
    log_likelihoods.append(log_likelihood)
    print(f"Log-Likelihood: {log_likelihood}")
    responsibilities = compute_responsibilities(X, weights, means, covariances)
    lower_bound = compute_lower_bound(X, responsibilities, weights, means, covariances)
    lower_bounds.append(lower_bound)
    print(f"Lower Bound (ELBO): {lower_bound}")
    weights, means, covariances = m_step(X, responsibilities)
    print("Updated weights in the M-step!")
    # compute log likelihood after M step
    log_likelihood = compute_log_likelihood(X, weights, means, covariances)
    log_likelihoods.append(log_likelihood)
    print(f"new Loglikelihood: {log_likelihood}")
    # compute lower bound with old responsibilities: -> Not tight!
    lower_bound = compute_lower_bound(X, responsibilities, weights, means, covariances)
    lower_bounds.append(lower_bound)
    print(f"Lower bound with old responsibilities: {lower_bound}")
    if e % 3 == 0:
        plot(X, weights, means, covariances)
```

## Plot of Log-Likelihood and Lower Bound over Time

```python
plt.plot(log_likelihoods, label="log-likelihood")
plt.plot(lower_bounds, label="lower bound")
plt.gca().set_xlabel("Steps (2 per epoch)")
plt.legend()
plt.show()
```
