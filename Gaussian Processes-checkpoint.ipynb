## Gaussian Processes


```python
import numpy as np 
import matplotlib.pyplot as plt

np.random.seed(0)

sigma_fct = 1        # standard deviation sigma from function noise 

x_train = np.load('x_train.npy')
y_train = np.load('y_train.npy')
x_plot = np.load('x_plot.npy')
y_plot = np.load('y_plot.npy')
```

```python
data_fig = plt.figure()
plt.plot(x_train, y_train, 'o')
plt.plot(x_plot, y_plot, 'r')
plt.show()
```

```python
def get_kernel_vec(X, x_prime, sigma):
    """
    @param X: training data matrix (N_train x d)
    @param x_prime: input to kernel vector (N_inputs x d)
    @param sigma: bandwidth of the kernel (scalar)
    @return: return kernel vector evaluated at each point of x_prime (N_train x N_inputs x d)
    """
    diff = np.zeros((X.shape[0], x_prime.shape[0], X.shape[1]))
    for i in range(X.shape[0]):
        diff [i, :, :] = X[i, :] - x_prime          # difference of each  training point with each input 
    return np.exp(-np.linalg.norm(diff, axis=2)**2/(2*sigma**2))
```

```python
def get_kernel_mat(X, sigma):
    """
    @param X: training data matrix (N_train, d)
    @sigma: bandwidth of the kernel(scalar)
    @return: the kernel matrix (N_train x N_train)
    """
    return get_kernel_vec(X, X, sigma)
```

In the following please implement the two functions "get_mean" and "get_variance" according to the equations in the slides:\
The predictive GP distribution is defined as\
Mean: $ \mu(\boldsymbol{x}^*) = \boldsymbol{k}_{\boldsymbol{x}^*}^T(\boldsymbol K +\sigma_y^2 \boldsymbol{I})^{-1}\boldsymbol{y} $ \
Variance: $ \sigma^2(\boldsymbol{x}^*) = \boldsymbol{k}^* +\sigma^2 - \boldsymbol{k}^T_{\boldsymbol{x}^*}(\boldsymbol{K}+\sigma^2 \boldsymbol{I})^{-1}\boldsymbol{k}_{\boldsymbol{x}^*} $ \
Don't forget to incorporate the $\lambda$ value into your equations according to the slides.

```python
def get_mean(x, x_train, y_train, bandwidth, inv_lamb_val, sigma_y):
    """
    @param x: input vector for which we want to get predictions (N_inputs x d)
    @param x_train: training data (N_train x d)
    @param bandwidth: bandwidth of the kernel
    @param inv_lamb_val: inverse precission value
    @param sigma_y: standard deviation of fucntion (not squared)
    @return: predicted values according to input x (N_inputs x d)
    """
    K = get_kernel_mat(x_train, bandwidth)*inv_lamb_val
    k_x = get_kernel_vec(x_train, x, bandwidth)*inv_lamb_val
    inv_mat_train_output = np.linalg.solve(K + sigma_y**2 * np.eye(K.shape[0]), y_train)
    return np.dot(k_x.T, inv_mat_train_output)
```

```python
def get_variance(x, x_train, bandwidth, inv_lamb_val, sigma_y):
    """
    @param x: input vector for which we want to get predictions (N_inputs x d)
    @param x_train: training data (N_train x d)
    @param bandwidth: bandwidth of the kernel
    @param inv_lamb_val: inverse precission value
    @param sigma_y: standard deviation of function (not squared)
    @return: predicted values according to input x (N_inputs x d)
    """
    K = get_kernel_mat(x_train, bandwidth)*inv_lamb_val
    k_x = get_kernel_vec(x_train, x, bandwidth)*inv_lamb_val
    inv_and_k_x = np.linalg.solve(K + sigma_y**2 * np.eye(K.shape[0]), k_x)
    k_star = get_kernel_vec(x, x, bandwidth)*inv_lamb_val
    return k_star + sigma_y**2 - np.dot(k_x.T, inv_and_k_x)
```

```python
bandwidth = 1                   # standard deviation of function noise (given)
sigma_y = sigma_fct
inv_lamb_val = 1000             # inverse lambda value -> equivalent to lambda = 1e-3
gp_fig = plt.figure()

# Let's go through the training data and add on training point to the system in each iteration and let's plot
# everything dynamically
x_dyn_train = []
y_dyn_train = []
for i in range(x_train.shape[0]):
    x_dyn_train.append(x_train[i])
    y_dyn_train.append(y_train[i])
    mean = get_mean(x_plot, np.array(x_dyn_train), np.array(y_dyn_train), bandwidth, inv_lamb_val, sigma_y)
    var = get_variance(x_plot, np.array(x_dyn_train), bandwidth, inv_lamb_val, sigma_y)
    if i % 5 == 0:
        plt.figure(gp_fig.number)
        gp_fig.clf()
        plt.plot(x_plot[:, 0], mean[:, 0])
        plt.fill_between(x_plot[:, 0], mean[:, 0] -2*np.sqrt(np.diag(var)), mean[:,0]+2*np.sqrt(np.diag(var)), alpha=0.2, edgecolor='r', facecolor='r')
        plt.plot(np.array(x_dyn_train), np.array(y_dyn_train), 'rx')
        plt.title('i='+ str(i+1))
        plt.pause(0.5)
    elif i == x_train.shape[0]-1:
        plt.figure(gp_fig.number)
        gp_fig.clf()
        plt.plot(x_plot[:, 0], mean[:, 0])
        plt.fill_between(x_plot[:, 0], mean[:, 0] -2*np.sqrt(np.diag(var)), mean[:,0]+2*np.sqrt(np.diag(var)), alpha=0.2, edgecolor='r', facecolor='r')
        plt.plot(np.array(x_dyn_train), np.array(y_dyn_train), 'rx')
        plt.title('i='+ str(i+1))
        plt.pause(0.5)

# now let's see the function approximation with all training data and compare to the ground truth function

mean = get_mean(x_plot, x_train, y_train, bandwidth, inv_lamb_val, sigma_y)
var = get_variance(x_plot, x_train, bandwidth, inv_lamb_val, sigma_y)

plt.figure()
plt.plot(x_plot[:, 0], mean[:, 0])
plt.fill_between(x_plot[:, 0], mean[:, 0] -2*np.sqrt(np.diag(var)), mean[:,0]+2*np.sqrt(np.diag(var)), alpha=0.2, edgecolor='r', facecolor='r')
plt.plot(np.array(x_train), np.array(y_train), 'rx')
plt.plot(x_plot, y_plot, 'g')

plt.legend(['mean prediction', 'training points', 'gt-function', '2 $\sigma$'])


```

In the last plots you can see how the GP is plotted for different iterations, where in each iteration a new training input is added in order to predict the data set ("x_plot"). 
We can see that the prediction is improved with increasing data points (->mean prediction gets better and the variance is reduced). \
However, you can also see for regions, where we don't have training data ([-2, -1]), the 95% intervall is high and the mean prediction is not precise. Neverthelles GP indicates that it is uncertain in this region, as the variance is higher in this region.
