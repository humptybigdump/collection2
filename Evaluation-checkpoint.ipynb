```python
def MT_reader(filename,dim):
    # open txt file
    with open(filename, "r") as f:
        content = f.readlines()
    f.close()
    # get info
    MT_list = {}
    for ci, info in enumerate([line.rstrip('\n') for line in content]):
        info = info.split()
        key = info[0]
        time = UTCDateTime(info[6])
        glob_loc = np.array([float(info[2]),float(info[3]),float(info[4])])
        mag = float(info[5])
        MT = np.array([float(info[7]),float(info[8]),float(info[9]),float(info[10]),float(info[11]),float(info[12])])
        if dim[0] <= glob_loc[0] <= dim[1] and dim[2] <= glob_loc[1] <= dim[3]:
            MT_list[key] = [time,glob_loc,mag,MT] 
    return MT_list


path2TextFiles = '/home/mlind/Python_Scripts/Final_Code/TextFiles/Karibik/'

filename = path2TextFiles+'Karibik_Moment_Tensor'
#MT_list = MT_reader(filename)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from py_src.util import Tape2M, M2Tape

def event_reader(fname,dim):
    # open txt file
    filename = '/home/mlind/Python_Scripts/Final_Code/TextFiles/Karibik/'+fname
    with open(filename, "r") as f:
        content = f.readlines()
    f.close()
    # get info
    event_list = {}
    for ci, info in enumerate([line.rstrip('\n') for line in content]):
        key = info.split()[0]
        time = str(UTCDateTime(info.split()[6]))
        glob_loc = [float(info.split()[2]),float(info.split()[3]),float(info.split()[4])]
        mag = float(info.split()[5])
        fm_temp = info.split()
        fm = [float(fm_temp[7]),float(fm_temp[8]),float(fm_temp[9]),float(fm_temp[10]),0.0]
        if dim[0] <= glob_loc[0] <= dim[1] and dim[2] <= glob_loc[1] <= dim[3]:
            event_list[key] = [time,glob_loc,mag,fm] 
    return event_list


def display_map_from_file(Results,MT_filename,map_dim):
    MT_list = MT_reader(MT_filename,map_dim)
    event_list_jiri = event_reader('Event_List_JIRI',map_dim)
    
    llcrnrlon = map_dim[2] # Longitude lower left corner
    llcrnrlat = map_dim[0] # Latitude lower left corner
    urcrnrlon = map_dim[3] # Longitude upper right corner
    urcrnrlat = map_dim[1] # Latitude upper right corner
    fsize = 15
    Mfac = 0.04
    
    fig = plt.figure(figsize=(25, 30), facecolor='w', edgecolor='k')
    ax = plt.subplot(1,1,1)

    m = Basemap(projection='cyl',# lon_0=-61.0, lat_0=15.0, 
                llcrnrlon=llcrnrlon, 
                urcrnrlon=urcrnrlon, 
                llcrnrlat=llcrnrlat,
                urcrnrlat=urcrnrlat, 
                resolution='i') 
    
    m.drawcoastlines()
    m.drawcountries()
    m.fillcontinents(color='wheat', lake_color='skyblue')
    # Draw a lon/lat grid (20 lines for an interval of one degree)
    m.drawparallels(np.linspace(llcrnrlat, urcrnrlat, 5), labels=[0, 1, 0, 0], fmt="%.2f",
                    dashes=[2, 2], rotation=0)
    m.drawmeridians(np.linspace(llcrnrlon, urcrnrlon, 5), labels=[0, 0, 0, 1], fmt="%.2f",
                    dashes=[2, 2])
    plt.yticks(fontsize=fsize)
    plt.xticks(fontsize=fsize)
    m.drawmapboundary(fill_color='skyblue')
    ax.set_title("Lesser Antilles",fontsize=fsize)
    
    for MT_id in MT_list:
        [time,loc,mag,MT] = MT_list[MT_id]
        x, y = m(loc[1], loc[0])
        z = loc[2]
        Mg_fac = Mfac*mag/6.5
        if UTCDateTime(2016, 10, 18, 22, 4) <= time <= UTCDateTime(2016, 10, 18, 22, 12):
            col = 'r'
        elif time < UTCDateTime(2020, 5, 15, 9, 3):
            col = 'b'
        else:
            col = 'b'
        b = beach(MT, xy=(x,y), width=Mg_fac, linewidth=1, facecolor=col, alpha=1.0, nofill=False)
        b.set_zorder(10)
        ax.add_collection(b)
        
        s1,d1,r1,c1,i1,M0 = M2Tape(np.asarray(MT))
        s2,d2,r2 = aux_plane(s1,d1,r1)
        
        b = beach([s1,d1,r1], xy=(x,y), width=Mg_fac, linewidth=1, facecolor='k', alpha=1.0, nofill=True)
        b.set_zorder(10)
        ax.add_collection(b)
        
        #plt.text(x,y,'-------'+str(int(loc[2])))
    
    for event_id in Results:
        fm = Results[event_id]['Result']['Result_DC']['result']['0'][0][1]
        loc = Results[event_id]['Container']['source']['F1_loc'][1]
        mag = Results[event_id]['Container']['source']['F1_loc'][2]

        if map_dim[0] <= loc[0] <= map_dim[1] and map_dim[2] <= loc[1] <= map_dim[3]:
        
            if event_id in ['ev113','ev146']:
                loc[0] += 0.05

            x, y = m(loc[1], loc[0])
            Mg_fac = Mfac*mag/6.5
            mt = Tape2M(fm[0],fm[1],fm[2],fm[3],fm[4],1)
            b = beach(mt, xy=(x,y), width=Mg_fac, linewidth=1, facecolor='r', alpha=1.0, nofill=False)
            b.set_zorder(10)
            ax.add_collection(b)

            plt.text(x,y,event_id)

            print(event_id,fm[:3],loc)
        
    
    for event_id in event_list_jiri:
        fm = event_list_jiri[event_id][3]
        loc = event_list_jiri[event_id][1]
        mag = event_list_jiri[event_id][2]
        x, y = m(loc[1], loc[0])
        Mg_fac = Mfac*mag/6.5
        mt = Tape2M(fm[0],fm[1],fm[2],0.*fm[3],0.*fm[4],1)
        b = beach(mt, xy=(x,y), width=Mg_fac, linewidth=1, facecolor='g', alpha=1.0, nofill=False)
        b.set_zorder(10)
        ax.add_collection(b)
        
        #plt.text(x,y,event_id)
    
    plt.show()
    
    
MT_filename = path2TextFiles+'Karibik_Moment_Tensor'
minlat, maxlat = 14.5, 18.5
minlon, maxlon = -62.38,-60.
map_dim = [minlat, maxlat, minlon, maxlon]
#map_dim = [17.40,17.60,-61.2,-60.9]
display_map_from_file(Results,MT_filename,map_dim)  
```

```python
from py_src.util import Tape2M
from obspy.imaging.beachball import beach, aux_plane

def get_P(Result):
    pdf = []
    nl2 = []
    for fi in range(len(Result['result']['0'])):
        pdf.append(Result['result']['0'][fi][0])
        temp = Result['result']['0'][fi][7]
        if temp >= 1.:
            nl2.append(1.0)
        else:
            nl2.append(temp)
    pdf = np.asarray(pdf)    
    nl2 = np.asarray(nl2)
    VR = 1.-nl2
    cdf = pdf/np.sum(pdf)
    
    P = []
    for ii in range(len(cdf)-1):
        pp = np.sum(cdf[:ii])
        P.append(pp)
    P = np.asarray(P)
    
    return P,cdf,nl2,VR

def get_fault(Result):
    
    FM_ref = Result['result']['0'][0][1]
    FM_ref_aux = aux_plane(FM_ref[0],FM_ref[1],FM_ref[2])
    MT_res = [FM_ref]
    for fmi in range(len(Result['result']['0'])):
        FM = Result['result']['0'][fmi][1]
        F1 = np.array([np.abs(FM[0]-FM_ref[0]),np.abs(FM[1]-FM_ref[1]),np.abs(FM[1]-FM_ref[1])])
        F2 = np.array([np.abs(FM[0]-FM_ref_aux[0]),np.abs(FM[1]-FM_ref_aux[1]),np.abs(FM[1]-FM_ref_aux[1])])
        
        if np.sum(F1) < np.sum(F2):
            MT_res.append(FM)
        else:
            FM_aux = aux_plane(FM[0],FM[1],FM[2])
            MT_res.append([FM_aux[0],FM_aux[1],FM_aux[2],FM[3],FM[4]])
        
    return MT_res



Table = {}
prob = 0.15
fpath = '/home/mlind/Python_Scripts/Final_Code/Misrotation/Paper_Figure/Results/'

for event_id in Results:
    
    Table[event_id] = {'Z':0,'R':0,'T':0}
    
    Container = Results[event_id]['Container']
    inv_pre = Results[event_id]['inv_pre']
    Result = Results[event_id]['Result']['Result_DC']
    
    Table[event_id]['event_time'] = Container['source']['F1_loc'][0]
    if np.abs(inv_pre['twind'][0][0]) >= 10:
        tup = inv_pre['twind'][0][0] - 30.
    Table[event_id]['event_time_update'] = round(tup,2)
    Table[event_id]['event_time_update_uncertainty'] = round(inv_pre['twind'][0][1],2)
    Table[event_id]['lat'] = Container['source']['F1_loc'][1][0]
    Table[event_id]['lon'] = Container['source']['F1_loc'][1][1]
    Table[event_id]['depth'] = Container['source']['F1_loc'][1][2]
    Table[event_id]['event_model_depth'] = Container['source']['event_depth']
    Table[event_id]['F_max'] = Container['preprocessing']['filter']['freqmin']
    Table[event_id]['F_min'] = Container['preprocessing']['filter']['freqmax']#['fcut'][0]
    
    trace_selection = Container['network']['trace_selection']
    for stat_id in trace_selection:
        for ci in trace_selection[stat_id]:
            Table[event_id][ci] += 1    
    
    P, cdf, nl2, VR = get_P(Result)
    nP = int(len(Result['result']['0'])*prob)
    Table[event_id]['Probability'] = round(100*P[nP],1)

    
    FM = get_fault(Result)
    strike,dip,rake,clvd,iso = [],[],[],[],[]
    Mw,NL2 = [], []
    for fi in range(nP):
        fm = FM[fi]
        strike.append(fm[0])
        dip.append(fm[1])
        rake.append(fm[2])
        clvd.append(fm[3])
        iso.append(fm[4])
        
        for comp in Result['result']['0'][fi][5]:
            for fbi in Result['result']['0'][fi][5][comp]:
                mw = Result['result']['0'][fi][5][comp][fbi][1]
                if mw == mw:
                    Mw.append(mw)
                nl2 = Result['result']['0'][fi][7]
                if nl2 == nl2: 
                    NL2.append(nl2)
        
    #'''   
    ax = plt.subplot(111)
    for fi in range(nP):
        fm = FM[fi]
        
        if fi == 0:
            nofill = False
        else:
            nofill = True
        
        try:
            mt = Tape2M(fm[0],fm[1],fm[2],fm[3],fm[4],1)
            b = beach(mt, width=200, xy=(0, 0),linewidth=1, 
                          facecolor='b',edgecolor='k' , alpha=1.0, nofill=nofill)
            ax.add_collection(b)
        except:
            continue
        ax.set_aspect('equal')
        ax.set_xlim([-101, 101])
        ax.set_ylim([-101, 101])
        plt.axis('off')
    b_filename = fpath+event_id+'.png'
    plt.savefig(b_filename,bbox_inches='tight',transparent=False,pad_inches=0)
    Table[event_id]['beachball'] = b_filename 
    #'''
    
        
    MTA = [np.mean(np.asarray(strike)),np.mean(np.asarray(dip)),np.mean(np.asarray(rake)),
           np.mean(np.asarray(clvd)),np.mean(np.asarray(iso))]
    MTB = aux_plane(MTA[0],MTA[1],MTA[2])
        
    Table[event_id]['Strike_N1'] = int(MTA[0])
    Table[event_id]['Strike_N2'] = int(MTB[0])
    Table[event_id]['Strike_std'] = int(np.std(np.asarray(strike)))
    
    Table[event_id]['Dip_N1'] = int(MTA[1])
    Table[event_id]['Dip_N2'] = int(MTB[1])
    Table[event_id]['Dip_std'] = int(np.std(np.asarray(dip)))
    
    Table[event_id]['Rake_N1'] = int(MTA[2])
    Table[event_id]['Rake_N2'] = int(MTB[2])
    Table[event_id]['Rake_std'] = int(np.std(np.asarray(rake)))
    
    Table[event_id]['CLVD'] = int(MTA[3])
    Table[event_id]['CLVD_std'] = int(100*np.std(np.asarray(clvd))/30.)
    
    Table[event_id]['ISO'] = int(MTA[4])
    Table[event_id]['ISO_std'] = int(100*np.std(np.asarray(iso))/90.)
    
    Table[event_id]['Mw'] = round(np.mean(np.asarray(Mw)),2)
    Table[event_id]['Mw_std'] = round(np.std(np.asarray(Mw)),2)
    
    VR = 1.-np.asarray(NL2)
    Table[event_id]['VR'] = round(np.mean(VR),2)
    Table[event_id]['VR_std'] = round(np.std(VR),2)
```
